# Making simple CNN as first pass

```python
#images are 150x150 pixels and 3 for the RGB color channels
img_input = layers.Input(shape=(150,150,3))

#First Conv extracts 16 filters, 3x3 ea followed by max-pooling layer
x = layers.Conv2D(16,3, activation='relu')(img_input)
x = layers.MaxPooling2D(2)(x)

#Second Conv extracts 32 filters, 3x3 ea followed by max-pooling layer
x = layers.Conv2D(32,3, activation='relu')(x)
x = layers.MaxPooling2D(2)(x)

#Third Conv extrats 64 filters, 3x3 ea followed by max-pooling layer
x = layers.Conv2D(64,3, activation='relu')(x)
x = layers.MaxPooling2D(2)(x)

#Flatten feature map to 1-dim to add fully connected layer
x = layers.Flatten()(x)

#Fully connected layer with 512 hidden units
x = layers.Dense(512, activation='relu')(x)

#Create output layer with 6 nodes and softmax activation for prediction
output = layers.Dense(6, activation='softmax')(x)

#creating model:
#  input = input feature map
#  output = input feature map + stacked conv/max-pooling layers + softmax output
model = Model(img_input, output)

model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer=optimizers.RMSprop(lr=.001),
             metrics=[metrics.categorical_accuracy])
```

```python
#Building generators to yield batches of 20 images of size 150x150 pixels and labels
#  also normalizing pixel values from [0,255] to [0,1]

train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

#flow training images in batches of 20 using train_datagen generator
train_generator = train_datagen.flow_from_directory(
    train_dir, #Source directory
    target_size=(150,150), #resizing all images to 150x150 pixels
    batch_size=20,
    class_mode='categorical',
    shuffle=True #choosing random training images
)

#flow validation images in batches of 20 using test_datagen generator
validation_generator = test_datagen.flow_from_directory(
    validation_dir,#Source directory
    target_size=(150,150), #resizing all images to 150x150 pixels
    batch_size=20,
    class_mode='categorical',
    shuffle=True #choosing random test images
)
```

```python
#training on only 4000 of the ~14000 total images and testing on 2000
model_history = model.fit_generator(
    train_generator,
    steps_per_epoch=200, #4000 images = 20 batch_size * 200 steps
    epochs=15,
    validation_data=validation_generator,
    validation_steps=100, #2000 images = 20 batch_size * 100 steps
    verbose=2
)
```

```python
def plot_model_history(model_history):
    acc = model_history.history['categorical_accuracy']
    val_acc = model_history.history['val_categorical_accuracy']
    
    loss = model_history.history['loss']
    val_loss = model_history.history['val_loss']
    
    num_epochs = range(len(acc))
    
    plt.plot(num_epochs,acc)
    plt.plot(num_epochs, val_acc)
    plt.title('Training and Validation accuracy')
    plt.legend(['Train','Test'], loc='upper left')
    
    plt.figure()
    plt.plot(num_epochs,loss)
    plt.plot(num_epochs,val_loss)
    plt.title('Training and Validation loss')
    plt.legend(['Train','Test'], loc='upper left')
    
```

```python
print(model_history.history.keys())
plot_model_history(model_history)
```

###### model is overfitting, so lets throw in some dropout regularization before prediction

```python
#images are 150x150 pixels and 3 for the RGB color channels
img_input = layers.Input(shape=(150,150,3))

#First Conv extracts 16 filters, 3x3 ea followed by max-pooling layer
x = layers.Conv2D(16,3, activation='relu')(img_input)
x = layers.MaxPooling2D(2)(x)

#Second Conv extracts 32 filters, 3x3 ea followed by max-pooling layer
x = layers.Conv2D(32,3, activation='relu')(x)
x = layers.MaxPooling2D(2)(x)

#Third Conv extrats 64 filters, 3x3 ea followed by max-pooling layer
x = layers.Conv2D(64,3, activation='relu')(x)
x = layers.MaxPooling2D(2)(x)

#Flatten feature map to 1-dim to add fully connected layer
x = layers.Flatten()(x)

#Fully connected layer with 512 hidden units
x = layers.Dense(512, activation='relu')(x)

#Dropout layer
x = layers.Dropout(0.5)(x)

#Create output layer with 6 nodes and softmax activation for prediction
output = layers.Dense(6, activation='softmax')(x)

#creating model:
#  input = input feature map
#  output = input feature map + stacked conv/max-pooling layers + softmax output
model = Model(img_input, output)

model.summary()

model.compile(loss='categorical_crossentropy',
             optimizer=optimizers.RMSprop(lr=.001),
             metrics=[metrics.categorical_accuracy])
```

```python
#training on only 2000 of the ~14000 total images and testing on 1000
model_history = model.fit_generator(
    train_generator,
    steps_per_epoch=100, #2000 images = 20 batch_size * 100 steps
    epochs=15,
    validation_data=validation_generator,
    validation_steps=50, #1000 images = 20 batch_size * 50 steps
    verbose=2
)
```

```python
plot_model_history(model_history)
```

##### Still overfitting, but lets try changing up the model architechure and see what happens

```python
#images are 150x150 pixels and 3 for the RGB color channels
img_input = layers.Input(shape=(150,150,3))

#First Conv extracts 200 filters, 3x3 ea
x = layers.Conv2D(200,3, activation='relu')(img_input)

#Second Conv extracts 180 filters, 3x3 ea
x = layers.Conv2D(180,3, activation='relu')(img_input)

#max-pooling layer
x = layers.MaxPooling2D(5)(x)

#Third Conv extrats 180 filters, 3x3 ea
x = layers.Conv2D(180,3, activation='relu')(x)

#Fourth Conv extrats 140 filters, 3x3 ea
x = layers.Conv2D(140,3, activation='relu')(x)

#Fifth Conv extrats 100 filters, 3x3 ea
x = layers.Conv2D(100,3, activation='relu')(x)

#Fifth Conv extrats 100 filters, 3x3 ea
x = layers.Conv2D(50,3, activation='relu')(x)

#max-pooling layer
x = layers.MaxPooling2D(5)(x)

#Flatten feature map to 1-dim to add fully connected layer
x = layers.Flatten()(x)

#Fully connected layer with 180 hidden units
x = layers.Dense(512, activation='relu')(x)

#Fully connected layer with 100 hidden units
x = layers.Dense(100, activation='relu')(x)

#Fully connected layer with 50 hidden units
x = layers.Dense(50, activation='relu')(x)

#Dropout layer
x = layers.Dropout(0.5)(x)

#Create output layer with 6 nodes and softmax activation for prediction
output = layers.Dense(6, activation='softmax')(x)

#creating model:
#  input = input feature map
#  output = input feature map + stacked conv/max-pooling layers + softmax output
model_v2 = Model(img_input, output)

model_v2.summary()

model_v2.compile(loss='categorical_crossentropy',
             optimizer=optimizers.RMSprop(lr=.001),
             metrics=[metrics.categorical_accuracy])
```

```python
start = time.time()
#training on only 2000 of the ~14000 total images and testing on 1000
model_v2_history = model_v2.fit_generator(
    train_generator,
    steps_per_epoch=100, #2000 images = 20 batch_size * 100 steps
    epochs=15,
    validation_data=validation_generator,
    validation_steps=50, #1000 images = 20 batch_size * 50 steps
    verbose=2
)
end = time.time()

print("Training time (mins) = ", (end - start)/60)
```

```python
plot_model_history(model_v2_history)
```

## Adding Google's inception v3 model to do the feature extraction

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3

local_weights_file = curr_dir + "/data/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5"
pre_trained_inception_model = InceptionV3(
    input_shape=(150,150,3),
    include_top=False,
    weights=None)

#making inception model weights not trainable
for layer in pre_trained_inception_model.layers:
    layer.trainable = False
```

```python
#Using the 'mixed7' layer of Inception to do the feature extraction
last_layer = pre_trained_inception_model.get_layer('mixed7')
last_output = last_layer.output

#Appending a fully connected classifier on top of last_output
x = layers.Flatten()(last_output)

#Fully connected layer with 1024 hidden units
x = layers.Dense(1024, activation='relu')

#Adding dropout layer
x = layers.Dropout(0.2)(x)

#final output softmax classifier
x = layers.Dense(6,activation='softmax')(x)

#configure and compile model
model_inception = Model(pre_trained_inception_model.input,x)
model_inception.compile(loss='categorical_crossentropy',
             optimizer=optimizers.RMSprop(lr=.001),
             metrics=[metrics.categorical_accuracy])
```

```python
start = time.time()
#training on only 2000 of the ~14000 total images and testing on 1000
model_inception_history = model_inception.fit_generator(
    train_generator,
    steps_per_epoch=100, #2000 images = 20 batch_size * 100 steps
    epochs=2,
    validation_data=validation_generator,
    validation_steps=50, #1000 images = 20 batch_size * 50 steps
    verbose=2
)
end = time.time()

print("Training time (mins) = ", (end - start)/60)
```
